# Add emissions to all the flights and trains

```python
import pandas as pd
```

#### Flights

```python
df_flights = pd.read_csv('../data/flight_connections_filtered.csv')
```

```python
CO2_per_km = 110 # g/km
average_people = 330 # people
CO2_per_liter = 3.15 # g/liter
takeoff_emissions_aircraft = 110000 # g
takeoff_emission_passenger = (takeoff_emissions_aircraft / average_people) * CO2_per_liter
```

```python
# add column for emissions per passenger
df_flights['emissions'] = df_flights['distance'] * CO2_per_km + takeoff_emission_passenger
```

```python
df_flights.head()
```

```python
df_flights.to_csv('../data/flight_connections_emissions.csv', index=False)
```

#### Trains

```python
df_trains = pd.read_csv('../data/train_connections_filtered.csv')
```

```python
train_CO2_per_km = 3.15 # g/km
```

```python
# add column for emissions per passenger
df_trains['emissions'] = df_flights['distance'] * train_CO2_per_km
```

```python
df_trains.head()
```

```python
# save to csv
df_trains.to_csv('../data/train_connections_emissions.csv', index=False)
```
