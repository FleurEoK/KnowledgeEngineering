# Preprocess airports 
Only include european airports with IATA codes and save to csv

```python
import pandas as pd
```

```python
# Read in the data
df_airports = pd.read_csv('../data/airports.csv', delimiter=';')
```

```python
countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic',
    'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary',
    'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta',
    'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia',
    'Spain', 'Sweden'
]
```

#### filter airports to only include european airports


```python
df_airports_EU = df_airports[df_airports['Country'].isin(countries)]
df_airports_EU
```

#### drop all rows where IATA is \N
Only take airports with IATA code

```python
df_airports_EU = df_airports_EU[df_airports_EU['IATA'] != '\\N']
```

```python
# save data to csv
df_airports_EU.to_csv('../data/airports_filtered.csv', index=False)
```
