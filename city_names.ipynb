# Make a csv with all city names

```python
import pandas as pd
# Read in the data
df_trains = pd.read_csv('../train_connections_filtered.csv')
df_flights = pd.read_csv('../flight_routes_filtered_with_distances.csv')
df_city_synonyms = pd.read_csv('../city_synonyms.csv')
```

```python
print(list(df_city_synonyms['synonymLabel']))
```

```python
# loop over all 'destination_city' and 'departure_city' in train and 'source_city' and 'dest_city' in flight data, and check if they are in the city synonyms list (if so, replace them with the city name in the list)
for i in range(len(df_trains)):
        for j in range(len(df_city_synonyms)):
            if df_trains['destination_city'][i] == df_city_synonyms['synonymLabel'][j]:
                df_trains['destination_city'][i] = df_city_synonyms['cityLabel'][j]
            if df_trains['departure_city'][i] == df_city_synonyms['synonymLabel'][j]:
                df_trains['departure_city'][i] = df_city_synonyms['cityLabel'][j]

for i in range(len(df_flights)):
        for j in range(len(df_city_synonyms)):
            if df_flights['source_city'][i] == df_city_synonyms['synonymLabel'][j]:
                df_flights['source_city'][i] = df_city_synonyms['cityLabel'][j]
            if df_flights['dest_city'][i] == df_city_synonyms['synonymLabel'][j]:
                df_flights['dest_city'][i] = df_city_synonyms['cityLabel'][j]
```

```python
cities = []
# add all cities all columns of interest without having duplicates
cities.extend(df_trains['destination_city'].unique())
cities.extend(df_trains['departure_city'].unique())
cities.extend(df_flights['source_city'].unique())
cities.extend(df_flights['dest_city'].unique())

# remove duplicates
cities = list(set(cities))
```

```python
# orderthem alphabetically
cities.sort()
cities
```

```python
# save them to a csv file with header 'city'
pd.DataFrame(cities, columns=['city']).to_csv('../cities.csv', index=False)

# save df_trains and df_flights to csv files
df_trains.to_csv('../train_connections_final.csv', index=False)
df_flights.to_csv('../flight_routes_final.csv', index=False)
```
