# Removing the unnamed column from the dataset

```python
import pandas as pd
# Read in the data
df_trains = pd.read_csv('../train_connections.csv')
df_flights = pd.read_csv('../flight_routes_with_distances.csv')
```

```python
df_trains.head()
```

```python
df_flights.head()
```

```python
# drop unnamed column
df_flights.drop('Unnamed: 0', axis=1, inplace=True)
df_trains.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# save the data
df_flights.to_csv('../flight_routes_filtered_with_distances.csv', index=False)
df_trains.to_csv('../train_connections_filtered.csv', index=False)
```
