# Preprocess routes
Include city and country of source and destination airports

```python
import pandas as pd
```

```python
# Read in the data
df_airports = pd.read_csv('../data/airports_filtered.csv')
df_routes = pd.read_csv('../data/routes.csv')
```

```python
countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic',
    'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary',
    'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta',
    'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia',
    'Spain', 'Sweden'
]
```

```python
country_codes = {
    'Austria': 'AT',
    'Belgium': 'BE',
    'Bulgaria': 'BG',
    'Croatia': 'HR',
    'Cyprus': 'CY',
    'Czech Republic': 'CZ',
    'Denmark': 'DK',
    'Estonia': 'EE',
    'Finland': 'FI',
    'France': 'FR',
    'Germany': 'DE',
    'Greece': 'GR',
    'Hungary': 'HU',
    'Ireland': 'IE',
    'Italy': 'IT',
    'Latvia': 'LV',
    'Lithuania': 'LT',
    'Luxembourg': 'LU',
    'Malta': 'MT',
    'Netherlands': 'NL',
    'Poland': 'PL',
    'Portugal': 'PT',
    'Romania': 'RO',
    'Slovakia': 'SK',
    'Slovenia': 'SI',
    'Spain': 'ES',
    'Sweden': 'SE'
}
```

```python
# rename column ' destination apirport' to 'destination airport'
df_routes.rename(columns={' destination apirport': 'destination airport'}, inplace=True)
```

```python
# make a dictionary with IATA as key and country as value
iata_country = {}
iata_city = {}
iata_longitude = {}
iata_latitude = {}
for index, row in df_airports.iterrows():
    iata_country[row['IATA']] = row['Country']
    iata_city[row['IATA']] = row['City']
    iata_longitude[row['IATA']] = row['Longitude']
    iata_latitude[row['IATA']] = row['Latitude']
```

```python
# filter routes to only include routes between countries in the EU
iatas = df_airports['IATA'].tolist()
df_routes = df_routes[df_routes['destination airport'].isin(iatas)]
df_routes = df_routes[df_routes[' source airport'].isin(iatas)]
```

```python
df_routes
```

```python
df_routes.columns
```

```python
# loop over all routes and add the country of the destination airport and the country of the source airport as columns
source_longitude = []
source_latitude = []
dest_longitude = []
dest_latitude = []
dest_city = []
source_city = []
dest_country = []
source_country = []
dest_country_code = []
source_country_code = []

for index, row in df_routes.iterrows():
    dest_city.append(iata_city[row['destination airport']])
    source_city.append(iata_city[row[' source airport']])
    dest_country.append(iata_country[row['destination airport']])
    source_country.append(iata_country[row[' source airport']])
    dest_country_code.append(country_codes[iata_country[row['destination airport']]])
    source_country_code.append(country_codes[iata_country[row[' source airport']]])
    dest_longitude.append(iata_longitude[row['destination airport']])
    source_longitude.append(iata_longitude[row[' source airport']])
    dest_latitude.append(iata_latitude[row['destination airport']])
    source_latitude.append(iata_latitude[row[' source airport']])

df_routes['source_longitude'] = source_longitude
df_routes['source_latitude'] = source_latitude
df_routes['source_city'] = source_city
df_routes['source_country'] = source_country
df_routes['source_country_code'] = source_country_code
df_routes['dest_longitude'] = dest_longitude
df_routes['dest_latitude'] = dest_latitude
df_routes['dest_city'] = dest_city
df_routes['dest_country'] = dest_country
df_routes['dest_country_code'] = dest_country_code
```
